## 1. Kütüphanelerin Yüklenmesi

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Histogramın Çizilmesi

```python
# Gauss eğrilerinin parametreleri
mu1 = 20
sigma1 = 5
mu2 = 30
sigma2 = 10
mu3 = 40
sigma3 = 8

# Veri örnekleri oluşturma
x = mu1 + (sigma1 * np.random.randn(10000))
y = mu2 + (sigma2 * np.random.randn(10000))
z = mu3 + (sigma3 * np.random.randn(10000))

# Histogramları çizme
n, bins, patches = plt.hist(x, 100, density = 1, facecolor = 'b', alpha = 0.25, label = 'Eğri 1')
n, bins, patches = plt.hist(y, 100, density = 1, facecolor = 'g', alpha = 0.25, label = 'Eğri 2')
n, bins, patches = plt.hist(z, 100, density = 1, facecolor = 'r', alpha = 0.25, label = 'Eğri 3')

# Eksen ve Başlık ayarları
plt.xlabel('Veriler')
plt.ylabel('Olasılıklar')
plt.title(r'Üç Farklı Gauss Eğrisi: $\mu_1 = 20,\ \mu_2 = 30,\ \mu_3 = 40$')
plt.annotate('G.E. Tepe Noktası', xy=(20,0.085), xytext=(30,0.09),
            arrowprops=dict(facecolor='black', shrink=0.1, alpha=0.25, connectionstyle="angle3"))

plt.text(3, 0.07, r'$\mu_1=20,\ \sigma_1=5$')
plt.text(22, 0.05, r'$\mu_2=30,\ \sigma_2=10$')
plt.text(45, 0.04, r'$\mu_3=40,\ \sigma_3=8$')
plt.axis([0,65,0,0.095])
plt.legend()
plt.grid(True)
plt.show()
```

## 3. Sonuç

#### Gauss Eğrilerinin Parametreleri:
   
   
Üç farklı Gauss eğrisi için ortalama (mu) ve standart sapma (sigma) değerlerini belirliyoruz. Bu değerler, her bir eğrinin istatistiksel özelliklerini tanımlıyor.

#### Veri Örnekleri Oluşturma: 

    
Numpy kütüphanesini kullanarak her bir Gauss eğrisi için veri örnekleri oluşturuyoruz. np.random.randn() fonksiyonu, standart normal dağılıma (ortalama 0, standart sapma 1) sahip rastgele sayılar üretir. Böylece sayıları veri noktalarına dönüştürerek istenilen Gauss eğrisini oluşturabiliriz.

#### Histogramları Çizme: 

    
Her bir Gauss eğrisi için plt.hist() fonksiyonu ile veri örnekleri histogram olarak çiziliyor.


#### Eksen ve Başlık Ayarları: 

    
Eksen etiketleri, başlık ve diğer görsel öğeler ekleniyor. Ayrıca 'plt.annotate()' fonksiyonu ile ok ekleyerek bir açıklama yapıyoruz.


#### Açıklamalar ve Eksen Sınırları: 

    
Açıklamaları ve eksen sınırlarını belirliyoruz.


#### Görselin Gösterilmesi:

    
plt.show() komutu ile oluşturulan grafiği görüntülüyoruz.

r'...' Gösterimi: 'r' ifadesi, bir "raw string" yani ham dizedir. Bu ifade, metin içindeki özel karakterlerin (örneğin, \) kaçış dizilerini yok saymasını sağlar. Özellikle matematiksel ifadelerde, LaTeX benzeri formatları kullanmak için kullanışlıdır. Örneğin, r'$\mu_1=20,\ \mu_2=30$' ifadesi, şeklinde bir matematiksel ifade oluştururken kaçış dizilerini önlemek için kullanılır.
